```python
import numpy as np
import pickle
import pytzer as pz
from matplotlib import pyplot as plt
%matplotlib notebook

with open('pickles/fortest_sim.pkl','rb') as f:
    crp94,F = pickle.load(f)
crp94['lnacfPM_sd'] = np.std(F['ln_acfPM'], axis=1)
crp94['sqtot'] = np.sqrt(crp94.tot)

with open('pickles/h2so4_prop_eye2.pkl','rb') as f:
    acrp94,UlnacfPM,UlnacfPM_var,UacfPM,UacfPM_var = pickle.load(f)
acrp94['sqtot'] = np.sqrt(acrp94.tot)
```

```python
fig,ax = plt.subplots(1,1)

L = crp94.temp == 273.15
crp94[L].plot.scatter('sqtot','lnacfPM_sd', ax=ax, c='b', label='273.15 K', alpha=0.5, s=20)
ax.scatter(acrp94.sqtot[L],np.sqrt(UlnacfPM_var[L]), c='b', alpha=0.2, s=20)

L = crp94.temp == 298.15
crp94[L].plot.scatter('sqtot','lnacfPM_sd', ax=ax, c='g', label='298.15 K', alpha=0.5, s=20)
ax.scatter(acrp94.sqtot[L],np.sqrt(UlnacfPM_var[L]), c='g', alpha=0.2, s=20)

L = crp94.temp == 323.15
crp94[L].plot.scatter('sqtot','lnacfPM_sd', ax=ax, c='orange', label='323.15 K', alpha=0.5, s=20)
ax.scatter(acrp94.sqtot[L],np.sqrt(UlnacfPM_var[L]), c='orange', alpha=0.2, s=20)

ax.grid(alpha=0.5)

ax.set_xlabel('(TSO$_4$ / mol/kg)$^{1/2}$')
ax.set_ylabel('$\sigma$ (ln $\gamma_\pm$)');
```
